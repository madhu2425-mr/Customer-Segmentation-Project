```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1. Generate Sample Customer Data (or load your own using pd.read_csv('your_data.csv'))
np.random.seed(42)
data = {
    'CustomerID': range(1, 201),
    'Age': np.random.randint(18, 70, size=200),
    'Annual_Income_k': np.random.randint(15, 130, size=200),
    'Spending_Score': np.random.randint(1, 100, size=200)
}
df = pd.DataFrame(data)

print("--- First 5 Rows of the Dataset ---")
print(df.head(), "\n")

# 2. Select Features for Clustering
# We use Annual Income and Spending Score for this demographic/behavioral segmentation
X = df[['Annual_Income_k', 'Spending_Score']]

# 3. Feature Scaling (Crucial for distance-based algorithms like K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Find the Optimal Number of Clusters using the Elbow Method
inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Curve to find the best 'K'
plt.figure(figsize=(8, 4))
plt.plot(K_range, inertia, marker='o', linestyle='--', color='b')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia / WCSS')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()

# 5. Apply K-Means with the Optimal Number of Clusters (Let's assume k=5 based on standard customer behavior patterns)
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_transform(X_scaled).argmin(axis=1) # Assign cluster labels

# 6. Analyze and Visualize the Segments
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Annual_Income_k',
    y='Spending_Score',
    hue='Cluster',
    data=df,
    palette='Set1',
    s=100,
    alpha=0.8
)
plt.title('Customer Segments Based on Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster/Segment')
plt.grid(True)
plt.show()

# Print out the profile/characteristics of each segment
print("--- Cluster Profiles (Mean Values) ---")
cluster_profiles = df.groupby('Cluster')[['Age', 'Annual_Income_k', 'Spending_Score']].mean()
print(cluster_profiles)
```
